```
gen = set(data["generation"])
mayor = 0
for g in gen:
  total = data[data.generation == g].sum()
  if mayor < total["suicides_no"]:
    mayor = total["suicides_no"]
    generacion = g
  print("Generacion =", g, ", Total de suicidios =",total["suicides_no"])

print("\nGeneracion con mas suicidios =", generacion, ", con =", mayor, "suicidios")
```

```
sexo = set(data["sex"])
mayor = 0
totalSuicidios = data[["suicides_no"]].sum() #Total de suicidios
for s in sexo:
  total = data[data.sex == s].sum()
  porc = float((total["suicides_no"]*100)/totalSuicidios)
  if mayor < total["suicides_no"]:
    mayor = total["suicides_no"]
    sexom = s
  print("Sexo =", s, ", total de suicidios =",total["suicides_no"], ", Porcentaje = %", porc)

print("\nSexo con mas suicidios =", sexom, ", con =", mayor, "suicidios") 
```

```
age = set(data["age"])
mayor = 0
totalSuicidios = data[["suicides_no"]].sum() #Total de suicidios
for s in age:
  total = data[data.age == s].sum()
  porc = float((total["suicides_no"]*100)/totalSuicidios)
  if mayor < total["suicides_no"]:
    mayor = total["suicides_no"]
    agem = s
  print("Rango de edad =", s, ", total de suicidios =",total["suicides_no"], ", Porcentaje = %", porc)

print("\nRango de edad con mas suicidios =", agem, ", con =", mayor, "suicidios") 
```

```
mayor = 0
totalSuicidios = data[["suicides_no"]].sum() #Total de suicidios

for i in range(1981, 2021, 10):
  print(f"DÃ©cada de {i - 1}s")
  total = data[data.year.between(i , i+9)].sum()
  porc = float((total["suicides_no"]*100)/totalSuicidios)
  if mayor < total["suicides_no"]:
    mayor = total["suicides_no"]
    yearm = i
  print("total de suicidios =",total["suicides_no"], ", Porcentaje = %", porc)

print("\nDÃ©cada con mas suicidios =", yearm-1, "'s, con =", mayor, "suicidios") 
```

```
country = set(data["country"])
mayor = 0
totalSuicidios = data[["suicides_no"]].sum() #Total de suicidios
auxl = []
for c in country:
  total = data[data.country == c].sum()
  porc = float((total["suicides_no"]*100)/totalSuicidios)
  auxl.append([total["suicides_no"], c, porc])

auxl.sort()
tam = len(auxl)

print("Los diez paÃ­ses con mayor nÃºmero de suicidios")
print("No. Suicidios\t\tPaÃ­s\t\tPorcentaje")
for x in range(10):
  for y in range(3):
    print(auxl[tam-(x+1)][y], end="\t\t")
    
  print()
```
